```python
#  silhoutte score to decide no of k

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Select the scaled numerical features for clustering
X_scaled = merged_df_scaled[['Annualized_Return', 'Annualized_Volatility', 'Log_ShareVolume', 'Log_Turnover']]

# Determine the range of k values to evaluate
k_range = range(2, 11) # Evaluating k from 2 to 10

# Lists to store scores
silhouette_scores = []

# Loop through the range of k values
for k in k_range:
    # Initialize KMeans with the current k
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Added n_init
    # Fit the KMeans model
    kmeans.fit(X_scaled)
    # Get the cluster labels
    labels = kmeans.labels_
    # Calculate the silhouette score
    score = silhouette_score(X_scaled, labels)
    # Append the score to the list
    silhouette_scores.append(score)
    print(f"Silhouette Score for k={k}: {score:.4f}")


# Plot the silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.xticks(k_range)
plt.grid(True)
plt.show()


```

```python

import matplotlib.pyplot as plt
# Use the elbow method to find the optimal number of clusters
inertia = []
k_range = range(2, 11)  # Evaluate k from 2 to 10

# Select the scaled numerical features for clustering
X_scaled = merged_df_scaled[['Annualized_Return', 'Annualized_Volatility', 'Log_ShareVolume', 'Log_Turnover']]


for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Added n_init
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_range)
plt.grid(True)
plt.show()

```

```python
#Hierarchical clustering

import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# Select the scaled numerical features for clustering (same as KMeans)
X_scaled = merged_df_scaled[['Annualized_Return', 'Annualized_Volatility', 'Log_ShareVolume', 'Log_Turnover']]

# Perform hierarchical clustering using Ward's method
# Ward's method minimizes the variance of the clusters being merged.
linked = linkage(X_scaled, method='ward')

# Plot the dendrogram
plt.figure(figsize=(18, 8))
dendrogram(linked,
           orientation='top',
           labels=merged_df_scaled['Company'].tolist(),
           distance_sort='descending',
           show_leaf_counts=False) # Set to True to see leaf counts at merges
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Company Index or Company Name')
plt.ylabel('Distance')
plt.show()

```

```python
from scipy.cluster.hierarchy import linkage, fcluster

# X =  scaled feature set
linked = linkage(X_scaled , method='ward')

# Cut dendrogram at a height (t = 25) and form flat clusters
cluster_labels = fcluster(linked, t=25, criterion='distance')

# Number of unique clusters
num_clusters = len(set(cluster_labels))
print("Number of clusters:", num_clusters)
```

```python
import numpy as np
import pandas as pd

# 1. Define Annualised Sharpe Ratio Function
def sharpe_ratio(weekly_returns, risk_free_rate_annual=0.0894):
    mean_weekly = np.mean(weekly_returns)
    std_weekly = np.std(weekly_returns)

    # Annualise mean and std deviation
    annualised_return = mean_weekly * 52
    annualised_volatility = std_weekly * np.sqrt(52)

    # Calculate Sharpe Ratio safely
    if annualised_volatility != 0:
        sharpe = (annualised_return - risk_free_rate_annual) / annualised_volatility
    else:
        sharpe = 0

    return sharpe

# 2. Calculate Sharpe Ratios for each company in cleaned_df
sharpe_ratios = []

for i in range(len(cleaned_df)):
    returns = cleaned_df.iloc[i].values
    sharpe = sharpe_ratio(returns)
    sharpe_ratios.append(sharpe)

# 3. Create Final Cluster DataFrame
cluster_df = pd.DataFrame({
    'Company': cleaned_df.index,
    'Cluster': labels,   # Ensure 'labels' is defined from your clustering output
    'Sharpe Ratio': sharpe_ratios
})
```

```python
import pandas as pd
import numpy as np

# Portfolio construction

# Calculate total companies and companies per cluster
total_companies = len(merged_ts)
cluster_counts = merged_ts['Cluster'].value_counts()

# Calculate proportional slots
proportions = (cluster_counts / total_companies) * 6

# Convert to integer slots, ensuring at least 1 per cluster
slots = proportions.apply(np.floor).astype(int)
slots[slots < 1] = 1

# Adjust total slots to exactly 6 by adding or subtracting
while slots.sum() < 6:
    # Add to cluster with highest decimal remainder in proportion
    remainders = proportions - slots
    add_cluster = remainders.idxmax()
    slots[add_cluster] += 1
while slots.sum() > 6:
    # Remove from cluster with lowest Sharpe company (to minimise impact)
    remove_cluster = slots.idxmax()
    if slots[remove_cluster] > 1:
        slots[remove_cluster] -= 1


```

```python
# Select companies based on slots
final_portfolio = []

for cluster, n_slots in slots.items():
    cluster_df = merged_ts[merged_ts['Cluster'] == cluster]

    # Drop rows with NaN Sharpe Ratio
    cluster_df = cluster_df.dropna(subset=['Sharpe Ratio'])

    # Sort by Sharpe Ratio descending
    cluster_df_sorted = cluster_df.sort_values(by='Sharpe Ratio', ascending=False)

    # Select top n_slots companies from this cluster
    selected = cluster_df_sorted.head(n_slots)

    # Append to final portfolio
    final_portfolio.append(selected)

# Concatenate all selected companies
final_portfolio_df = pd.concat(final_portfolio, ignore_index=True)

# Display final portfolio
print(final_portfolio_df[['Company', 'Cluster', 'Sharpe Ratio']])
```
